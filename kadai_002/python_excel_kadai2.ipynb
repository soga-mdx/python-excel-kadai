```python
import pandas as pd

data = {
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店']
}
df = pd.DataFrame(data)

df['売上'] = pd.to_numeric(df['売上'])

overall_average = df['売上'].mean()

df['全体平均売上'] = overall_average


def performance(sales_average):

    if sales_average >= overall_average + 50:
        return 'A'
    elif sales_average >= overall_average:
        return 'B'
    else:
        return 'C'

df['業績ランク'] = df['売上'].apply(performance)

df.to_excel('業績.xlsx', index=False)
```
